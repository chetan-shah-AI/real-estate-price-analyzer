```python
import sqlite3


conn = sqlite3.connect('example.db')

c = conn.cursor()

c.execute('''CREATE TABLE IF NOT EXISTS example_test
                (id INTEGER PRIMARY KEY, name TEXT, age TEXT)''')
conn.commit()

```
