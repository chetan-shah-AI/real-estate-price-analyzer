```python
import requests

url = "https://www.rightmove.co.uk/api/_search"

params = {
    "locationIdentifier": "REGION^789",
    "radius": "10.0",
    "minPrice": "50000",
    "maxPrice": "15000000",
    "minBedrooms": "0",
    "maxBedrooms": "10",
    "propertyTypes": "detached,semi-detached,terraced",
    "maxDaysSinceAdded": "1",
    "includeSSTC": "true",
    "numberOfPropertiesPerPage": "24",
    "index": "0",
    "sortType": "6",
    "channel": "BUY",
    "currencyCode": "GBP"
}

headers = {
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_15_7)",
    "Accept": "application/json, text/javascript, */*; q=0.01",
    "Referer": "https://www.rightmove.co.uk/",
    "Origin": "https://www.rightmove.co.uk",
    "X-Requested-With": "XMLHttpRequest"
}

res = requests.get(url, params=params, headers=headers)

print("Status:", res.status_code)
print("Content-Type:", res.headers.get("Content-Type"))
print(res.text[:500])  # DEBUG

# Only parse if valid JSON
if "application/json" in res.headers.get("Content-Type", ""):
    data = res.json()
    print(len(data.get("properties", [])))
else:
    print("❌ Not JSON — likely blocked")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
import time
import pandas as pd

# Setup driver
options = webdriver.ChromeOptions()
options.add_argument("--start-maximized")
options.add_argument("--disable-blink-features=AutomationControlled")

driver = webdriver.Chrome(
    service=Service(ChromeDriverManager().install()),
    options=options
)

url = "https://www.rightmove.co.uk/property-for-sale/find.html?searchLocation=Leicester%2C+Leicestershire&useLocationIdentifier=true&locationIdentifier=REGION%5E789&radius=10.0&minPrice=50000&maxPrice=15000000&minBedrooms=0&maxBedrooms=10&propertyTypes=detached%2Csemi-detached%2Cterraced&maxDaysSinceAdded=1&_includeSSTC=on"

driver.get(url)

# wait for page to load
time.sleep(5)

# Accept cookies if popup appears
try:
    driver.find_element(By.ID, "onetrust-accept-btn-handler").click()
    time.sleep(2)
except:
    pass

properties = []

cards = driver.find_elements(By.CSS_SELECTOR, "[data-testid='propertyCard']")

for card in cards:
    try:
        price = card.find_element(By.CSS_SELECTOR, "[data-testid='property-price']").text
    except:
        price = None

    try:
        address = card.find_element(By.CSS_SELECTOR, "[data-testid='property-address']").text
    except:
        address = None

    try:
        beds = card.find_element(By.CSS_SELECTOR, "[data-testid='property-beds']").text
    except:
        beds = None

    try:
        link = card.find_element(By.TAG_NAME, "a").get_attribute("href")
    except:
        link = None

    properties.append({
        "price": price,
        "address": address,
        "bedrooms": beds,
        "link": link
    })

driver.quit()

df = pd.DataFrame(properties)
print(df.head())
```
