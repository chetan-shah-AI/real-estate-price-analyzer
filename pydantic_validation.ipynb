```python
from pydantic import BaseModel

class User(BaseModel):
    id: int
    name: str
    age: int

user_data = {
    "id": 1,
    "name": "Alice",
    "age": 30
}

user = User(id=user_data["id"], name=user_data["name"], age=user_data["age"])
print(user)
```
