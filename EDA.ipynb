```python
def describe_city_data(city):
    df = pd.read_csv(f"CLEANED_DATA/rightmove/rm_properties_{city.lower()}.csv")
    print(f"Data for {city}:")
    print(df.info())
    display(df.describe())
    print("\n")

list_cities = ["Leicester", "Birmingham", "London", "Glasgow", "Edinburgh", "Newcastle", "Bristol", "Manchester", "Nottingham", "Sheffield", "Liverpool", "Cardiff", "Belfast", "Leeds", "Southampton"]
for city in list_cities:
    try:
        describe_city_data(city)
    except Exception as e:
        print(f"Error describing data for {city}: {e}")
# describe_city_data("Birmingham")
```
