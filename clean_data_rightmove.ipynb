```python
def clean_rightmove_data(file_path, output_path):
    df = pd.read_csv(file_path)
    df["price"] = df["price"].str.replace(r"\D", "", regex=True)
    df["area"] = df["area"].str.replace(r"\D", "", regex=True)
    df.to_csv(output_path, index=False)


clean_rightmove_data("RAW_DATA/rightmove/properties_with_bedrooms_auto_london.csv", "CLEANED_DATA/rightmove/rm_properties_london.csv")
```

## Clean Incosistent Data From Zoopla

```python
import pandas as pd

df_rm_london = pd.read_csv("RAW_DATA/zoopla/zoopla_properties_London.csv")
```

```python
df_rm_london
```

```python
import os 
os.makedirs("CLEANED_DATA/zoopla", exist_ok=True)

def clean_rightmove_data(file_path, output_path):
    df = pd.read_csv(file_path)
    df["price"] = df["price"].str.replace(r"\D", "", regex=True)
    df["bedrooms"] = df["bedrooms"].str.replace(r"\D", "", regex=True)
    df["bathrooms"] = df["bathrooms"].str.replace(r"\D", "", regex=True)
    df["area"] = df["area"].astype(str).str.replace(r"\D", "", regex=True)
    # print("Area cleaned")

    df.to_csv(output_path, index=False)


clean_rightmove_data("RAW_DATA/zoopla/zoopla_properties_London.csv", "CLEANED_DATA/zoopla/zoopla_properties_london.csv")
```

```python
list_cities = ["Leicester", "Birmingham", "London", "Glasgow", "Edinburgh", "Newcastle", "Bristol", "Manchester", "Notingham", "Sheffield", "Liverpool", "Cardiff", "Belfast", "Leeds", "Southampton"]

for city in list_cities:
    try:
        print(f"Cleaning data for {city}...")
        clean_rightmove_data(f"RAW_DATA/zoopla/zoopla_properties_{city}.csv", f"CLEANED_DATA/zoopla/zoopla_properties_{city.lower()}.csv")
    except Exception as e:
        print(f"Error cleaning data for {city}: {e}")
```
